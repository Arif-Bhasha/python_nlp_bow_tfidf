```python
stp=stopwords.words('english')
```

```python
def cleandf(x,PorterStemmer):
    lf=[]
    for sent in x.head(1000):
        l=[]
        ls=sent.lower()
        ls=emoji.demojize(ls)
        ls=re.sub("<.*?","",ls)
        ls=re.sub("https?://\S*|www\S*","",ls)
        ls=re.sub("[^a-zA-Z\s]","",ls)
        for word in word_tokenize(ls):
            if word in stp and len(word)<=2:
                pass
            else:
                stems=PorterStemmer().stem(word)
                l.append(stems)
        lf.append(" ".join(l))
```
